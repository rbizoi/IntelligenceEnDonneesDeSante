# L’opérateur INTERSECT

<img src="https://raw.githubusercontent.com/rbizoi/IntelligenceEnDonneesDeSante/main/InterrogationSQL/images/intersect.png" width="400">


```python
%%sql 
SELECT VILLE,                                        
       NOM_PRODUIT                                   
FROM CLIENTS CL                                      
     JOIN COMMANDES CO                               
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)        
     JOIN DETAILS_COMMANDES DC                       
        ON ( CO.NO_COMMANDE = DC.NO_COMMANDE)        
     JOIN PRODUITS PR                                
        ON ( DC.REF_PRODUIT = PR.REF_PRODUIT)        
INTERSECT                                            
SELECT VILLE, NOM_PRODUIT                            
FROM PRODUITS PR                                     
     JOIN FOURNISSEURS FR                            
        ON ( PR.NO_FOURNISSEUR = FR.NO_FOURNISSEUR )
```

# L’opérateur DIFFERENCE

<img src="https://raw.githubusercontent.com/rbizoi/IntelligenceEnDonneesDeSante/main/InterrogationSQL/images/difference.png" width="400">


```python
%%sql 
SELECT CO.NO_COMMANDE,                          
       CO.DATE_COMMANDE                         
FROM CLIENTS CL                                 
     JOIN COMMANDES CO                          
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)   
     JOIN DETAILS_COMMANDES DC                  
        ON ( CO.NO_COMMANDE = DC.NO_COMMANDE)   
     JOIN PRODUITS PR                           
        ON ( DC.REF_PRODUIT = PR.REF_PRODUIT)   
WHERE PR.CODE_CATEGORIE = 1                     
MINUS                                           
SELECT CO.NO_COMMANDE,                          
    CO.DATE_COMMANDE                            
FROM CLIENTS CL                                 
     JOIN COMMANDES CO                          
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)   
     JOIN DETAILS_COMMANDES DC                  
        ON ( CO.NO_COMMANDE = DC.NO_COMMANDE)   
     JOIN PRODUITS PR                           
        ON ( DC.REF_PRODUIT = PR.REF_PRODUIT)   
WHERE PR.CODE_CATEGORIE = 9                     
```

# Sous-requête monolignes

<img src="https://raw.githubusercontent.com/rbizoi/IntelligenceEnDonneesDeSante/main/InterrogationSQL/images/sr01.png" width="400">


```python
%%sql 
SELECT NOM_PRODUIT 
FROM PRODUITS 
WHERE UNITES_STOCK = (SELECT MAX(UNITES_STOCK) FROM PRODUITS)
```

```python
%%sql 
SELECT CL.SOCIETE,                                     
       CO.NO_COMMANDE,                                 
       COUNT( DC.REF_PRODUIT)                          
FROM CLIENTS CL                                        
     JOIN COMMANDES CO                                 
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)          
     JOIN DETAILS_COMMANDES DC                         
        ON ( CO.NO_COMMANDE = DC.NO_COMMANDE)          
GROUP BY CL.SOCIETE,                                   
         CO.NO_COMMANDE                                
HAVING COUNT( DC.REF_PRODUIT) >=                       
   ( SELECT COUNT( DC.REF_PRODUIT)                     
     FROM COMMANDES CO                                 
          JOIN DETAILS_COMMANDES DC                    
               ON ( CO.NO_COMMANDE = DC.NO_COMMANDE)   
     WHERE  CO.NO_COMMANDE = 224975)
```

# Sous-requête multilignes

<img src="https://raw.githubusercontent.com/rbizoi/IntelligenceEnDonneesDeSante/main/InterrogationSQL/images/sr02.png" width="400">


```python
%%sql 
SELECT SOCIETE,                                                
       VILLE,                                                  
       PAYS                                                    
FROM CLIENTS                                                   
WHERE  CODE_CLIENT IN (                                        
               SELECT CO.CODE_CLIENT                           
               FROM COMMANDES CO                               
                    JOIN EMPLOYES EM                           
                         ON ( EM.NO_EMPLOYE = CO.NO_EMPLOYE)   
               WHERE EM.REND_COMPTE = 11)
```

# Sous-requête synchronisée

<img src="https://raw.githubusercontent.com/rbizoi/IntelligenceEnDonneesDeSante/main/InterrogationSQL/images/sr03.png" width="400">


```python
%%sql 
SELECT 
    CODE_CATEGORIE "Cat",                           
    NOM_PRODUIT,                                    
    UNITES_STOCK  "Stock",                          
    PRIX_UNITAIRE "Prix"                            
FROM PRODUITS P1                                       
WHERE UNITES_STOCK >(                                  
         SELECT AVG(UNITES_STOCK)*2                    
         FROM PRODUITS P2                              
         WHERE P2.CODE_CATEGORIE = P1.CODE_CATEGORIE)
```
